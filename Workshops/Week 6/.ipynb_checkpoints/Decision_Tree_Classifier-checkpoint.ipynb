### 3. For Humidity Feature 

```python
# Here total number of samples = 14
# Humidity (High) = 7 (3 yes, 4 no)
# Humidity (Normal) = 7 (6 yes, 1 no)

Gini_for_High = calc_Gini(3/7,4/7)
Gini_for_Normal = calc_Gini(6/7,1/7)
Gini_for_High, Gini_for_Normal
```

```python
#Calculation of weighted gini impurity for Humidity
weighted_Gini_Humidity = (7/14)*Gini_for_High + (7/14)*Gini_for_Normal
weighted_Gini_Humidity
```

```python
# Calculation of information gain 
# Gain_Humidity = parent_gini -weighted_Gini_Humidity
# print("Gain of Humidity:", Gain_Humidity)
```

### 4. For Wind Feature 

```python
# Here total number of samples = 14
# Wind (Strong) = 6 (3 yes, 3 no)
# Wind (Weak) = 8 (6 yes, 2 no)

Gini_for_Strong = calc_Gini(3/6,3/6)
Gini_for_Weak = calc_Gini(6/8,2/8)
Gini_for_Strong, Gini_for_Weak
```

```python
#Calculation of weighted gini impurity for Humidity
weighted_Gini_Wind = (6/14)*Gini_for_Strong + (8/14)*Gini_for_Weak
weighted_Gini_Wind
```
